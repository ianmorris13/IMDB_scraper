```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use(['dark_background'])
plt.suptitle('figure title', color='w')
```

```python
df = pd.read_csv('results.csv')
df.head()
```

```python
lol = df['movie_or_TV_name'].tolist()
lol = [l.split(',') for l in lol]
cmlptL = [el for l in lol for el in l]
cmlptL[:20]
```

```python
from collections import Counter
dicCnt = dict(Counter(cmlptL))
dicCnt
```

```python
df = pd.DataFrame([dicCnt])
df = df.transpose()
df = df.reset_index()
df.head()
```

```python
df = df.rename(columns={"index": "TV Show/ Movie", 0: "Number of Shared Actors"})
df.head()
```

```python
df = df.sort_values(by=['Number of Shared Actors'], ascending=False)
df = df.reset_index(drop=True)
df
```

```python
df[:20].plot.bar(x=r'TV Show/ Movie', y= 'Number of Shared Actors', rot=45, figsize = (20,5))
```
